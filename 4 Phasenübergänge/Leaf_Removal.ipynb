```python
import networkx as nx
import matplotlib.pyplot as plt
options = {
    # "font_size": 36,
    # "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    # "linewidths": 5,
    # "width": 5,
}
```

# Graph c=2

```python
G = nx.read_adjlist("Graph_c2.adjlist")
nx.write_adjlist(G, "Graph_c2.adjlist")

# nx.draw_networkx(G, **options)
ax = plt.gca()
ax.margins(0.20)

leaves = list((v for v, d in G.degree() if d == 0 or d == 1))
while len(leaves) != 0:
    for leaf in leaves:
        G.remove_node(leaf)
    leaves = list((v for v, d in G.degree() if d == 0 or d == 1))
    
plt.axis("off")
plt.show()

# After removal
fig, ax = plt.subplots(figsize=(10,10))
nx.draw_networkx(G, **options, ax=ax, pos=nx.circular_layout(G))
plt.axis("off")
plt.show()

core = list(n for n in G.nodes())
nx.write_adjlist(G, "Graph_c2_solved.adjlist")
print(core)
```
