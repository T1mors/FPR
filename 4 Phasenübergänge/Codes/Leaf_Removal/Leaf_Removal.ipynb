### Bei dieser Implementierung wird nicht versucht ein Vertex-Cover über den Leaf-Removal-Algorithmus zu bestimmen. Stattdessen wird der Algorithmus nur benötigt, um den core (Kern) des Graphen zu erhalten.

Usage: In Graph_cXXX.adjlist, whereby XXX is 2, 2.72 or 4, the edges are stored. So for example "20 16" means there is a connection between node 20 and 16. In Graph_cXXX_solved.adjlist the connections of the core are stored.

```python
import networkx as nx
import matplotlib.pyplot as plt
options = {
    "node_color": "white",
    "edgecolors": "black",
}
```

# Graph c=2

```python
G = nx.read_adjlist("Graph_c2.adjlist")
nx.write_adjlist(G, "Graph_c2.adjlist")

fig, ax = plt.subplots(figsize=(10,10))
nx.draw_networkx(G, **options, ax=ax, pos=nx.circular_layout(G))
plt.axis("off")
plt.show()


# Compute Core
leaves = list((v for v, d in G.degree() if d == 0 or d == 1))
while len(leaves) != 0:
    for leaf in leaves:
        edg = list(G.edges(G.neighbors(leaf)))
        G.remove_edges_from(edg)
        G.remove_node(leaf)
    leaves = list((v for v, d in G.degree() if d == 0 or d == 1))

# After removal
fig, ax = plt.subplots(figsize=(10,10))
nx.draw_networkx(G, **options, ax=ax, pos=nx.circular_layout(G))
#nx.draw_networkx(G, **options, ax=ax, pos =nx.kamada_kawai_layout(G))
plt.axis("off")
plt.show()

core = list(n for n in G.nodes())
nx.write_adjlist(G, "Graph_c2_solved.adjlist")
print("Core:", core)
#fig.savefig("LR_Graph_c2_Circular.png",dpi=400,bbox_inches = 'tight')
```
