```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from os.path import normpath, dirname
import json
import git

label_size = 20
plt.rcParams.update({'font.size': label_size, 'legend.title_fontsize': 18, 'legend.fontsize': 18, "axes.labelsize": label_size, "xtick.labelsize": label_size, 
"ytick.labelsize": label_size, "axes.labelpad": 4, "lines.markersize": 10, "lines.linewidth": 2})

# This will bind the paths from localSettings.json to objects in a class named paths
# usage: paths.<name>
class JSONObject:
    def __init__(self, dict):
        vars(self).update(dict)
with open(normpath(os.path.dirname(os.getcwd())+'/localSettings.json'),'r') as f: path_data = f.read()
paths = json.loads(path_data, object_hook=JSONObject)

# Headers start with #. So read this line and give the <header> vector to pd.read_csv(names=header)
def get_header(filename):
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith('#'):
                header = line
            else:
                break #stop when there are no more #

        header = header[1:].strip().split()
    return header
    
# /bin/python3 -m pip install ipykernel -U --user --force-reinstall
# /bin/python3.6 -m pip install numpy -U --user 
# in case this might help someone else, for me a workaround was to download the package from anaconda and install it with '--use-local'
# conda install -n myEnv --use-local myPackageFromAnaconda.tar.bz2
# https://github.com/conda/conda/issues/7245
# little disclaimer - any subsequent installation was screwed by the package
```
