# 4.2 Überprüfung des Gesetzes von Malus

```python
winkel = np.arange(-90,91,10)
winkel_fine = np.arange(-90,90+0.01,0.01)
spannung = [0.03, 0.34, 0.93, 1.76, 2.78, 3.89, 4.84, 5.57, 6.02, 6.10, 5.78, 5.13, 4.22, 3.14, 2.00, 1.07, 0.41, 0.05, 0.05]
spannung_slm = [5.97, 5.66, 5.08, 4.23, 3.21, 2.17, 1.24, 0.56, 0.14, 0.09, 0.44, 1.09, 1.96, 3.00, 4.09, 4.95, 5.59, 5.93, 5.91]
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
ax.scatter(winkel, spannung, fc="None", ec="C0", s=60, label="Messung ohne SLM")
ax.scatter(winkel, spannung_slm, fc="None", ec="C1", s=60, label="Messung mit SLM")


U_0 = max(spannung)
ax.plot(winkel_fine, U_0*np.cos(np.radians(winkel_fine))**2, label="Theorie")

ax.set_xlim(-95, 95)
ax.set_xticks(np.arange(-90,91,15))
ax.set_xlabel(r"$\alpha$ / $\degree$")
ax.set_ylabel("U / $V$")

ax.legend()
fig.tight_layout()
fig.savefig("Abb/Sec4_2_Gesetz_Malus.png")
```

# 4.3 SLM Grayscreen

```python
GW = [25, 70, 125, 150, 185, 235]
U_max = np.array([5.85, 5.72, 5.35, 5.42, 5.68, 6.02])
UMAX = max(U_max)
U_max = U_max/UMAX
Winkel_max = np.array([-7, -18, -41, -58, -83, 86])
U_min = np.array([0.01, 0.02, 0.26, 0.41, 0.11, 0.06])
U_min = U_min/UMAX
Winkel_min = [81, 73, 48, 33, 9, -2]
```

```python
# plt.rcParams.update({'xaxis.labellocation': "right"})
fig, ax = plt.subplots(1,1, figsize=(10,10))

for num, (gw, i, j, w) in enumerate(zip(GW, U_max, U_min, Winkel_max)):
    e = Ellipse(xy=(0,0), width=2*i, height=2*j, angle=w, label=f"{gw}")
    ax.add_patch(e)
    e.set_facecolor("None")
    e.set_edgecolor(f"C{num}")

ax.set_xlabel("U / V")
ax.set_ylabel("U / V", rotation = 0)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.xaxis.set_label_coords(1.05, 0.57)
ax.yaxis.set_label_coords(0.55, 1.03)

for spine in ax.spines.values():  # adjust spines on last active axis 
    spine.set_position(("data", 0))

ax.legend(title="Grauwert", loc=(0.7,0.55))
plt.tight_layout()
plt.savefig("Abb/Sec4_3Grayscreen.png")
```
